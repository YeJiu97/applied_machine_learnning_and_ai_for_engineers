# 二分类模型

定义：二分类模型是一种机器学习模型，用于将数据划分为两个类别。例如，在医学领域，判断一个患者是否患有某种疾病；在金融领域，判断一个客户是否会发生违约等。

常见算法：

1. **逻辑回归（Logistic Regression）**
   - 原理：逻辑回归是基于逻辑函数（Sigmoid函数）来预测数据属于某个类别的概率。其数学表达式为：
     $$
     P(y = 1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_nx_n)}}
     $$
     其中，$x_1, x_2, \dots, x_n$ 是输入特征，$\beta_0, \beta_1, \dots, \beta_n$ 是模型参数，$y$ 是目标变量（0或1）。Sigmoid函数的输出范围是(0, 1)，可以将这个输出解释为数据属于类别1的概率。如果概率大于某个阈值（如0.5），就将数据归为类别1，否则归为类别0。
   - 优点：模型简单，易于理解和解释，计算效率高，适合处理线性可分的数据。
   - 缺点：对于非线性问题的分类效果较差，且对异常值比较敏感。

2. **支持向量机（SVM）**
   - 原理：SVM的目标是找到一个最优的超平面，将两个类别的数据分开，使得两个类别之间的间隔最大。对于线性可分的情况，超平面的方程可以表示为：
     $$
     w \cdot x + b = 0
     $$
     其中，$w$ 是法向量，$b$ 是截距。对于非线性可分的情况，可以通过核函数（如高斯核、多项式核等）将数据映射到高维空间，然后在高维空间中寻找最优超平面。
   - 优点：在高维空间中表现良好，对于小样本数据集有较好的泛化能力，能够处理非线性问题。
   - 缺点：对于大规模数据集训练速度较慢，核函数的选择和参数调整比较复杂。


# 多分类模型

定义：多分类模型用于将数据划分为多个类别（三个或更多）。例如，在图像识别中，识别一张图片是猫、狗还是鸟；在文本分类中，将新闻文章归类为体育、财经、娱乐等不同类别。

常见算法

1. **决策树（Decision Tree）**
   - 原理：决策树通过一系列的规则（通常是基于特征的阈值判断）来对数据进行分类。它从根节点开始，根据特征的值选择分支，直到到达叶子节点，叶子节点表示最终的类别。例如，对于一个简单的多分类问题，决策树可能先判断特征$A$的值是否大于某个阈值，如果大于就进入一个分支，否则进入另一个分支，然后在每个分支中继续根据其他特征进行判断，最终到达叶子节点确定类别。
   - 优点：模型直观易懂，可以生成规则，方便解释。能够处理各种类型的数据，包括数值型和分类型数据。
   - 缺点：容易过拟合，特别是在数据特征较多且数据量较少的情况下。对于连续型数据，划分点的选择可能比较敏感。
2. **K近邻（K - Nearest Neighbors，KNN）**
   - 原理：KNN是一种基于实例的学习方法。对于一个新的数据点，它会找到训练数据集中与该数据点距离最近的K个点（这些点称为近邻点）。然后根据这K个近邻点的类别，通过投票等方式来确定新数据点的类别。距离的计算通常采用欧氏距离、曼哈顿距离等。例如，对于一个多分类问题，假设K = 3，新数据点的三个最近邻点中有两个属于类别A，一个属于类别B，那么新数据点就被归为类别A。
   - 优点：简单易实现，不需要训练模型，对于多分类问题表现良好。
   - 缺点：计算量较大，特别是当数据集很大时，需要计算新数据点与所有训练数据点的距离。对噪声数据和不相关特征比较敏感。

# 评估指标

分类模型的评估指标用于衡量模型的性能，不同的指标从不同的角度反映模型的优劣。以下是一些常用的评估指标：

1. **准确率（Accuracy）**
   - 定义：准确率是正确预测的样本数与总样本数的比值。其公式为：
     $$
     \text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
     $$
     其中，TP表示真正例（正确预测为正类的样本数），TN表示真负例（正确预测为负类的样本数），FP表示假正例（错误预测为正类的样本数），FN表示假负例（错误预测为负类的样本数）。
   - 优点：直观反映模型的整体性能。
   - 缺点：对于不平衡数据集（正负类样本数差异较大），准确率可能具有误导性。

2. **精确率（Precision）**
   - 定义：精确率是正确预测为正类的样本数与预测为正类的总样本数的比值。其公式为：
     $$
     \text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
     $$
   - 优点：反映模型预测正类的准确性。
   - 缺点：只关注正类的预测情况，忽略负类的预测情况。

3. **召回率（Recall）**
   - 定义：召回率是正确预测为正类的样本数与实际为正类的总样本数的比值。其公式为：
     $$
     \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
     $$
   - 优点：反映模型对正类的识别能力。
   - 缺点：只关注正类的识别情况，忽略负类的识别情况。

4. **F1分数（F1 Score）**
   - 定义：F1分数是精确率和召回率的调和平均数。其公式为：
     $$
     \text{F1 Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
     $$
   - 优点：综合考虑精确率和召回率，适用于不平衡数据集。
   - 缺点：对于多分类问题，需要计算每个类别的F1分数并进行平均。

5. **ROC曲线（Receiver Operating Characteristic Curve）**
   - 定义：ROC曲线是通过改变分类阈值，绘制出的模型的真阳性率（TPR）与假阳性率（FPR）的关系曲线。其中，TPR = Recall，FPR = FP / (FP + TN)。
   - 优点：直观反映模型在不同分类阈值下的性能，适用于比较不同模型的优劣。
   - 缺点：对于不平衡数据集，ROC曲线可能具有误导性。

6. **AUC值（Area Under the ROC Curve）**
   - 定义：AUC值是ROC曲线下的面积，其值范围在0到1之间。AUC值越大，表示模型的性能越好。
   - 优点：综合反映模型的性能，适用于比较不同模型的优劣。
   - 缺点：对于不平衡数据集，AUC值可能具有误导性。

7. **混淆矩阵（Confusion Matrix）**
   - 定义：混淆矩阵是一个表格，用于显示模型的预测结果与实际结果之间的关系。其行表示实际类别，列表示预测类别，每个单元格的值表示对应类别的样本数。
   - 优点：直观反映模型的分类情况，可以计算出各种评估指标。
   - 缺点：对于多分类问题，混淆矩阵可能比较复杂。

8. **Kappa系数（Cohen's Kappa）**
   - 定义：Kappa系数是衡量两个分类结果之间一致性的指标，其值范围在-1到1之间。Kappa系数越大，表示一致性越高。
   - 优点：考虑随机一致性，适用于比较不同模型的优劣。
   - 缺点：对于不平衡数据集，Kappa系数可能具有误导性。

9. **对数损失（Logarithmic Loss）**
   - 定义：对数损失是衡量模型预测概率与实际结果之间差异的指标，其值越小，表示模型的性能越好。
   - 优点：考虑预测概率，适用于比较不同模型的优劣。
   - 缺点：对于不平衡数据集，对数损失可能具有误导性。

10. **Brier分数（Brier Score）**
    - 定义：Brier分数是衡量模型预测概率与实际结果之间差异的指标，其值越小，表示模型的性能越好。
    - 优点：考虑预测概率，适用于比较不同模型的优劣。
    - 缺点：对于不平衡数据集，Brier分数可能具有误导性。

综上所述，分类模型的评估指标有多种，不同的指标从不同的角度反映模型的优劣。在实际应用中，需要根据具体问题和数据集的特点，选择合适的评估指标来衡量模型的性能。

# 二分类建模

## 泰坦尼克号

导入泰坦尼克号数据集：

```python
df = pd.read_csv("./../data/titanic/train.csv")
df.head()
```

查看是否有缺失值：

```python
df.info()
```

可以发现Cabin这里存在着大量的缺失值，不过关系不大，因为这一列不会被用到。

而age这一列则存在着一些缺失值，我们可以使用平均值来进行填充。

或者我们也可以直接删除带有缺失值的行：

```python
df = df[['Survived', 'Age', 'Sex', 'Pclass']]
df = pd.get_dummies(df, columns = ['Sex', 'Pclass'])

df.dropna(inplace = True)
df.head()
```

接着是针对训练集和测试集来拆分数据集：

```python
from sklearn.model_selection import train_test_split

x = df.drop('Survived', axis = 1)
y = df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

在上述代码中，很重要的一个参数是 stratify = y，这意味着在进行随机分割的时候，会在train和test中包含同等比例的正负样本。

接着创建一个逻辑回归模型：

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

接着我们还可以使用交叉验证法来对模型进行评分：

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, x, y, cv = 5).mean()
```

这个值其实更加的准确。

显示一下混淆矩阵：

```python
%matplotlib inline

from sklearn.metrics import ConfusionMatrixDisplay as cmd

cmd.from_estimator(model, x_test, y_test,
                  display_labels = ['Perished', 'Survived'],
                  cmap = 'Blues', xticks_rotation = 'vertical')
```

可以发现，在预测乘客会不幸的时候，模型的预测能力更加的准确。这是因为数据集中乘客死亡的样本本身就是远多于乘客幸存的样本。

可以使用其他的评估指标来进行模型评估：

```python
from sklearn.metrics import precision_score, recall_score

y_pred = model.predict(x_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

sensitivity = recall
specificity = recall_score(y_test, y_pred, pos_label = 0)
```

```python
print(f'Precision:{precision}')
print(f'Recall: {recall}')
print(f'Sensitivity: {sensitivity}')
print(f'Specificity: {specificity}')
```

这个逻辑回归模型在泰坦尼克号数据集上的表现呈现出一定的平衡性，但存在一些明显的优缺点。模型的精确率达到了 **85.42%**，表明在预测乘客“生存”时具有较高的准确性，即大多数被预测为生存的乘客确实生存了。然而，召回率只有 **70.69%**，这意味着模型在识别所有实际生存的乘客方面表现稍弱，仍有约 **29.31%** 的生存乘客被错误地预测为未生存。同时，模型的特异性较高，达到 **91.76%**，说明在预测乘客“未生存”时表现良好，能够有效地识别出大部分实际未生存的乘客。总体来看，该模型在处理“未生存”类别时表现较为出色，但在全面捕捉“生存”类别方面还有提升空间，可能需要进一步优化以提高对生存乘客的识别能力。

```python
# 进行预测

female = pd.DataFrame({'Age' : [30], 'Sex_female' : [1], "Sex_male" : [0], 'Pclass_1': [1], 'Pclass_2': [0], 'Pclass_3': [0]})
```

```python
# 预测一下
model.predict(female)[0]
```

预测能够幸存，但是我们需要考虑概率。

```python
probability = model.predict_proba(female)[0][1]
print(f"Probability of survival: {probability:.1%}")
```

## 信用卡检测

导入欧洲信用卡数据集：

```python
import pandas as pd
```

```python
score_card_df01 = pd.read_csv("./../data/creditcard_01.csv")
score_card_df02 = pd.read_csv("./../data/creditcard_02.csv")
```

```python
score_card_df01.shape, score_card_df02.shape
```

```python
credit_card_full = pd.concat([score_card_df01, score_card_df02], axis=0)
```

```python
credit_card_full.shape
```

首先是x特征：

```python
from sklearn.model_selection import train_test_split

x = credit_card_full.drop(['Time', 'Class'], axis = 1)
y = credit_card_full['Class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state = 0)
```

然后来训练一个逻辑回归划分不同类别：

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(random_state = 0, max_iter = 5000)
lr_model.fit(x_train, y_train)
```

默认的情况之下，max_iter的值是100，但是这对于上述的代码而言不够，所以需要指定max_iter = 5000。

查看一下模型的性能：

```python
%matplotlib inline
from sklearn.metrics import ConfusionMatrixDisplay as cmd

labels = ['Legitimate', 'Fraudulent']
cmd.from_estimator(lr_model, x_test, y_test, display_labels=labels, cmap='Blues', xticks_rotation='vertical')
```

这个逻辑回归模型正确的识别了56853笔交易为合法交易，但是又11笔交易被错误的归类为了欺诈性交易。

我们希望能够降低后面一个数字，这里尝试一下随机森林分类器。

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state = 0)
rf_model.fit(x_train, y_train)
```

查看结果：

```python
cmd.from_estimator(rf_model, x_test, y_test, display_labels = labels, cmap = 'Blues', xticks_rotation = 'vertical')
```

可以发现随机森林模型的表现更好很多。

这里还可以尝试使用梯度提升模型：

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm_model = GradientBoostingClassifier(random_state = 0)
gbm_model.fit(x_train, y_train)
```

```python
cmd.from_estimator(gbm_model, x_test, y_test, display_labels = labels, cmap = 'Blues', xticks_rotation = 'vertical')
```

最后决定使用随机森林模型。

```python
# 这里具体的查看一下
from sklearn.metrics import precision_score, recall_score

y_pred = rf_model.predict(x_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
sensitivity = recall
specificity = recall_score(y_test, y_pred, pos_label = 0)
```

```python
print(f'Precision:{precision}')
print(f'Recall: {recall}')
print(f'Sensitivity: {sensitivity}')
print(f'Specificity: {specificity}')
```

以下是对于信用卡欺诈预测结果的描述：

- **Precision（精确率）**：0.958904109589041，约95.89%。这意味着在模型预测为欺诈的交易中，有95.89%实际上是真正的欺诈交易。该指标反映了模型预测正例的准确性，即模型在识别欺诈交易时的可靠性较高。
- **Recall（召回率）**：0.7142857142857143，约71.43%。这表示在所有实际发生的欺诈交易中，模型成功识别出71.43%的欺诈案例。该指标反映了模型对实际欺诈交易的覆盖能力，即模型在所有欺诈交易中能够检测出的比例。
- **Sensitivity（敏感性）**：与Recall相同，为0.7142857142857143，约71.43%。它衡量了模型对正例（欺诈交易）的敏感程度，即正确识别出的欺诈交易占所有实际欺诈交易的比例。
- **Specificity（特异性）**：0.9999472425436128，约99.99%。这表示在所有实际非欺诈（正常）交易中，模型正确识别出99.99%的正常交易。该指标反映了模型对负例（正常交易）的识别能力，即模型在避免将正常交易误判为欺诈交易方面表现非常出色。

对于信用卡公司来说，他们通常更关注**Recall（召回率）** 和 **Specificity（特异性)** 这两个指标，但具体更关注哪一个可能取决于公司的业务策略和风险偏好。

- **Recall（召回率）**：信用卡公司希望尽可能多地检测出实际的欺诈交易，以减少因欺诈而导致的财务损失和客户不满。较高的召回率意味着模型能够覆盖更多的欺诈案例，从而降低漏报的风险。
- **Specificity（特异性）**：另一方面，信用卡公司也不希望将过多的正常交易误判为欺诈交易，因为这会导致客户体验下降（如交易被错误拒绝），并可能引发客户投诉和流失。较高的特异性确保了大多数正常交易能够顺利进行，减少了对客户正常使用的干扰。

在实际业务中，信用卡公司可能需要在这两个指标之间找到一个平衡点，以满足风险控制和客户满意度的双重目标。通常，信用卡公司会更倾向于较高的召回率，因为漏报的欺诈交易可能会带来更大的经济损失和声誉风险，而误判正常交易为欺诈虽然会影响客户体验，但可以通过后续的客户支持和补偿措施来缓解。
